Q) Write a program to compute the length of the hypotenuse (c) of a right triangle using Pythagoras
theorem

```python
import math

a = 3
b = 4

c = math.sqrt(a*a + b*b)
print("Hypotenuse:", c)
```

Q)Write a program to convert given seconds into hours, minutes and remaining second

```python
seconds = 3665

hours = seconds // 3600
minutes = (seconds % 3600) // 60
remaining_seconds = seconds % 60

print("Hours:", hours)
print("Minutes:", minutes)
print("Seconds:", remaining_seconds)
```

Q) Write a program to swap two numbers without taking additional variable

```python
a = 5
b = 10

a, b = b, a

print("a =", a)
print("b =", b)
```

Q) Write a program to find sum of first n natural numbers

Q) Write a program to print truth table for bitwise operators (&, | and ^ operators)

```python
a = 5
b = 3

print("a & b =", a & b)
print("a | b =", a | b)
print("a ^ b =", a ^ b)
```
