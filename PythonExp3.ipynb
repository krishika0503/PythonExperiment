Find a factorial of given number

```python
n = int(input("Enter number: "))
fact = 1

for i in range(1, n + 1):
    fact *= i

print("Factorial =", fact)
```

Check whether number is armstrong or not

```python
n = int(input("Enter number: "))
s = 0
t = n

while t > 0:
    d = t % 10
    s = s + d*d*d
    t = t // 10

if s == n:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

Print Fibonacci series up to given term.

```python
n = int(input("Enter number of terms: "))

a, b = 0, 1

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

Write a program to find if given number is prime number or not.

```python
n = int(input("Enter number: "))

if n <= 1:
    print("Not Prime")
else:
    for i in range(2, n):
        if n % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime Number")
```

Check whether the number is palindrome

```python
num = input("Enter a number:")

is_palindrome = True
length = len(num)

for i in range(length // 2):
    if num[i] != num[length - i - 1]:
        is_palindrome = False
        break

if is_palindrome:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

Write a program to print sum of digits.

```python
n = int(input("Enter number: "))
sum_digits = 0

while n != 0:
    sum_digits += n % 10
    n //= 10

print("Sum of digits =", sum_digits)
```

Count and print all numbers divisible by 5 or 7 between 1 to 100.

```python
count = 0

for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        print(i, end=" ")
        count += 1

print("\nTotal count =", count)
```

Write a program to print the following pattern

```python
n = 5

for i in range(n):

    # Left numbers
    for j in range(1, n - i + 1):
        print(j, end="")

    # Middle stars with space
    if i > 0:
        print(" ", end="")
        print("* " * i, end="")

    # Right numbers
    for j in range(n - i, 0, -1):
        if j != n - i + 1:
            print(j, end="")

    print()
```

Write a program to print the sum of the following series
1+ ½ + 1/3 + ¼ +….+1/n

```python
n = int(input("Enter value of n: "))
sum_series = 0

for i in range(1, n + 1):
    sum_series += 1 / i

print("Sum of series =", sum_series)
```
